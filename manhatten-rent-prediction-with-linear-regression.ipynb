# Rent Prediction in Manhatten

Using SciKit Learn to create a linear regression model to predict the possible rent of appartments

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

- This is the data set that is used to generate the predictions.

- y and x have been defined to train and test the model.

- All of the data set is being used to then make predictions

- This method could be tested and further improved and investigated

```python
streeteasy = pd.read_excel("../input/rent-costs-manhatten/ManhattenCosts.xlsx")
df = pd.DataFrame(streeteasy)

x = df[['bedrooms', 'bathrooms', 'size_sqft', 'min_to_subway', 'floor', 'building_age_yrs', 'no_fee', 'has_roofdeck', 'has_washer_dryer', 'has_doorman', 'has_elevator', 'has_dishwasher', 'has_patio', 'has_gym']]

y = df[['rent']]
```

2. Use train_test_split from SciKit learn to split the data 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state = 6)
```

3. Create a linear regression model and fit the train data

```python
mlr = LinearRegression()
mlr.fit(x_train, y_train) 
```

4. Predict y values from the test data 

```python
y_predict = mlr.predict(x_test)
```

Here an example prediction was made.

```python
sonny_apartment = [[3, 2, 800, 10, 5, 30, 1, 0, 1, 0, 0, 1, 1, 0]]
 
predict = mlr.predict(sonny_apartment)
 
print("Predicted rent: $%.2f" % predict)
```
