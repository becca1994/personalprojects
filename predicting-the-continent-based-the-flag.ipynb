The attached dataset has information on the countries of the world inlcuding flag features and other attributes. Decision trees have been made and optimised to predict the continent a country is on based on its flag. 


```python
#packages
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# load data
flags = pd.read_csv("/kaggle/input/flags/flags.csv", header = 0)
# creating data and labels
labels = flags[["Landmass"]]
data = flags[["Red", "Green", "Blue", "Gold", "White", "Black", "Orange", "Circles", "Crosses", 
              "Saltires", "Quarters", "Sunstars", "Crescent", "Triangle", "Bars", "Stripes", 
              "Colors"]]

train_data, test_data, train_labels, test_labels = train_test_split(data, labels, random_state=1)

# make and test the model
#for n in range(24,36,2):
#  for i in range(6,9):
#    for d in range(25,35,2):
#        tree = DecisionTreeClassifier(random_state = i, max_depth = d, max_leaf_nodes = n)
#        tree.fit(train_data, train_labels)
#        score = tree.score(test_data, test_labels)
#        print(n,i,d, score)
```

 max_leaf_nodes = 26, random_state = 6, max_depth = 16 would be the best tree available with these parameters. 

```python
tree = DecisionTreeClassifier(random_state = 6, max_depth = 16, max_leaf_nodes = 26)
tree.fit(train_data, train_labels)
score = tree.score(test_data, test_labels)
score
```
